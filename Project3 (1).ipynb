```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading the Dataset


```python
# Load the dataset
dataset_path = "/content/drive/MyDrive/Work/Projects/hotel_booking.csv"
data = pd.read_csv(dataset_path)
```

# Data Cleaning

```python
data.drop(['company','agent'], axis=1, inplace=True)
```

```python
data.isnull().sum()
```

```python
data['children'].median()
```

```python
#Filling the null values with the median value.
data['children'].fillna(data['children'].median(), inplace=True)
```

```python
data.isnull().sum()
```

```python
data['country'].mode()
```

```python
data.loc[data['country'] == 'PRT', 'country'].count()
```

```python
data['country'].fillna(data['country'].mode()[0], inplace=True)
```

```python
data.isnull().sum()
```

```python
print(data.columns)
```

# Exploratory Data Analysis


```python
# Plotting a graph to show a histogram for all the columns in the dataset
data.hist(figsize=(15,15))
plt.show()
```

```python
# Visualizing the bookings by pie chart.
ax = data['hotel'].value_counts().plot.pie(explode=[0.05, 0.05], autopct='%1.1f%%', shadow=True, figsize=(8,6), fontsize=16, colors=['pink','orange'])
ax.set_title('Percentage Share of Different Hotel Types ', fontsize=15, fontweight='bold')
```

```python
ax=sns.countplot(x='is_canceled', data=data, palette='Set2')
ax.set_title('Cancelled Bookings',fontsize = 16, fontweight='bold')
ax.set_xlabel('Cancelled',fontsize = 14)
ax.set_ylabel('Reservation Count',fontsize = 14)
for rect in ax.patches:
   ax.text (rect.get_x() + rect.get_width()  / 2,rect.get_height()+ 0.20,rect.get_height(),horizontalalignment='center', fontsize = 15)
```

```python
# Customer Preferences
sns.countplot(x='hotel', hue='is_canceled', data=data, palette='Set2')
plt.title('Hotel Type vs Cancellation')
plt.xlabel('Hotel Type')
plt.ylabel('Number of Bookings')
plt.legend(title='Cancellation')
plt.show()
```

```python
# Cancellation Trends by Market Segment
plt.figure(figsize=(9, 6))
sns.countplot(x='market_segment', hue='is_canceled', data=data, palette='Set2')
plt.title('Cancellation Trends by Market Segment')
plt.xlabel('Market Segment')
plt.ylabel('Number of Bookings')
plt.legend(title='Cancellation')
plt.xticks(rotation=45)
plt.show()
```

```python
# Cancellation Trends by Customer Type
sns.countplot(x='customer_type', hue='is_canceled', data=data, palette='Set2')
plt.title('Cancellation Trends by Customer Type')
plt.xlabel('Customer Type')
plt.ylabel('Number of Bookings')
plt.legend(title='Cancellation')
plt.show()
```

```python
# Seasonality Analysis
sns.countplot(x='arrival_date_month', data=data, order=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])
plt.title('Number of Bookings by Month')
plt.xlabel('Month')
plt.ylabel('Number of Bookings')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.countplot(x='arrival_date_month', hue='hotel', data=data, palette='Set2')
plt.title('Number of Bookings per Month by Hotel Type')
plt.show()
```

```python
# Lead Time Analysis
plt.figure(figsize=(10, 6))
sns.histplot(x='lead_time', data=data, bins=30, kde=True)
plt.title('Lead Time Distribution')
plt.xlabel('Lead Time (days)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Cancellation Trends
sns.countplot(x='deposit_type', hue='is_canceled', data=data, palette='Set2')
plt.title('Deposit Type vs Cancellation')
plt.xlabel('Deposit Type')
plt.ylabel('Number of Bookings')
plt.legend(title='Cancellation')
plt.show()
```

```python
# Distribution of Booking Types across Hotel Types
plt.figure(figsize=(8, 6))
sns.countplot(x='hotel', hue='reservation_status', data=data, palette='Set2')
plt.title('Booking Types Distribution across Hotel Types')
plt.xlabel('Hotel Type')
plt.ylabel('Number of Bookings')
plt.legend(title='Booking Type')
plt.show()
```

```python
# Distribution of Lead Time across Market Segments
plt.figure(figsize=(10, 6))
sns.boxplot(x='market_segment', y='lead_time', data=data, palette='Set2')
plt.title('Lead Time Distribution across Market Segments')
plt.xlabel('Market Segment')
plt.ylabel('Lead Time (days)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribution of Lead Time across Customer Types
plt.figure(figsize=(8, 6))
sns.boxplot(x='customer_type', y='lead_time', data=data, palette='Set2')
plt.title('Lead Time Distribution across Customer Types')
plt.xlabel('Customer Type')
plt.ylabel('Lead Time (days)')
plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.lineplot(x='arrival_date_month', y='is_canceled', hue='hotel', data=data.groupby(['arrival_date_month', 'hotel'])['is_canceled'].mean().reset_index(), palette='Set2')
plt.title('Cancellation Rate by Month and Hotel Type')
plt.xlabel('Month')
plt.ylabel('Cancellation Rate')
plt.show()
```

```python
numerical_data = data.select_dtypes(include=[np.number])
corr = numerical_data.corr()
plt.figure(figsize=(15, 10))
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap', fontsize=20)
plt.show()
```
